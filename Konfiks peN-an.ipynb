# A. Konfiks  peN-an
Konfik peN-an (disebut juga peng--an) dibentuk dari prefiks peN- dan sufiks -an. Konfik peN-an memiliki bentuk-bentuk yang disebut alomorf.

### (peng-an)
1. Konfik peN-an berbentuk peng-an jika bertemu kata berfonem awal vokal (a, i, u, e, o) atau konsonan /g/, /h/, /k/, /kh/ dan /q/ sedangkan fonem /k/ luluh.
Contoh:
peN-{ambil alih}-an → pengambilalihan
peN-{kirim}-an → pengiriman

### (pem-an)
2. Konfik peN-an berbentuk pem-an jika bertemu kata berfonem awal /b/, /p/, /f/, dan /v/ sedangkan fonem /p/ luluh.
Contoh:
peN-{baca}-an → pembacaan
peN-{putus}-an → pemutusan

### (peny-an)
3. Konfik peN-an berbentuk peny-an jika bertemu kata berfonem awal /s/.
Contoh:
peN-{salur}-an → penyaluran
peN-{serah terima}-an → penyerahterimaan

### (pen-an)
4. Konfik peN-an berbentuk pen-an jika bertemu kata berfonem awal /d/, /t/, /c/, /j/, /z/, dan /sy/ sedangkan fonem /t/ luluh.
Contoh:
peN-{dapat}-an → pendapatan
peN-{temu}-an → penemuan

### (pe-an)
5. Konfik peN-an berbentuk pe-an jika bertemu kata berfonem awal /l/, /r/, /w/, /y/, dan nasal (n, ng, ny, m).
Contoh:
peN-{rasa}-an → perasaan
peN-{warna}-an → pewarnaan

### (penge-an)
6. Konfik peN-an berbentuk penge-an jika bertemu kata dasar bersuku satu.
Contoh:
peN-{bom}-an → pengeboman
peN-{tes}-an → pengetesan


# MESIN VALIDASI KATA BERKONFIKS peN-an


```python
masukkan = raw_input('Masukkan kata :')
# print('halo', masukkan)
print('\n\n')
peLs = ['l','r','w','y','m','n']
pengLs = ['a','i','u','e','o','g','h','k','q']
pemLs = ['b', 'p', 'f', 'v']
penLs = ['d', 't', 'c', 'j', 'z']
vocalLs = ['a','i','u','e','o']



#cek kata keseluruhan, apakah ada di df..
if df.loc[(df.Kata == masukkan)].empty:
    #pe-an
    if (masukkan[:2] == 'pe') & (masukkan[-2:] == 'an') & ( not df.loc[(df.Kata == masukkan[2:-2])].empty ):   
        for alfabet in peLs:
            if masukkan[2]==alfabet:
                print('VALID!!')
                print('Konfiks : pe-an\n' )
                print(df.loc[(df.Kata == masukkan[2:-2])])
    #peng-an
    elif (masukkan[:4] == 'peng') & (masukkan[-2:] == 'an') & ( not df.loc[(df.Kata == masukkan[4:-2])].empty ):   
        for alfabet in pengLs:
            if masukkan[4]==alfabet:
                print('VALID!!') #k = luluh
                print('Konfiks : peng-an\n' )
                print(df.loc[(df.Kata == masukkan[4:-2])])
    #pem-an
    elif (masukkan[:3] == 'pem') & (masukkan[-2:] == 'an') & ( not df.loc[(df.Kata == masukkan[3:-2])].empty ):   
        for alfabet in pemLs:
            if masukkan[3] == alfabet:
                print('VALID!!') #k = luluh
                print('Konfiks : pem-an\n' )
                print(df.loc[(df.Kata == masukkan[3:-2])])
    #pen-an
    elif (masukkan[:3] == 'pen') & (masukkan[-2:] == 'an') & ( not df.loc[(df.Kata == masukkan[3:-2])].empty ):   
        for alfabet in penLs:
            if masukkan[3] == alfabet:
                print('VALID!!') #k = luluh
                print('Konfiks : pen-an\n' )
                print(df.loc[(df.Kata == masukkan[3:-2])])
    #peny-an
    elif (masukkan[:4] == 'peny') & (masukkan[-2:] == 'an') & ( not df.loc[(df.Kata == ('s'+masukkan[4:-2]))].empty ):   
        print('VALID!!') #k = luluh
        print('Konfiks : peny-an\n' )
        print('s' + df.loc[(df.Kata == masukkan[4:-2])])
    #penge-an
    elif (masukkan[:5] == 'penge') & (masukkan[-2:] == 'an') & ( not df.loc[(df.Kata == masukkan[5:-2])].empty ):   
        i = 0
        for alfabet in vocalLs:
            for font in masukkan[5:-2]:
                if alfabet == font:
                    i=i+1
        if i==1:
            print('VALID!!') #k = luluh
            print('Konfiks : penge-an\n' )
            print(df.loc[(df.Kata == masukkan[5:-2])])            
    else:
        print('Tidak VALID!')
                
else :
    print( masukkan ,"tidak memiliki konfiks", masukkan," merupakan kata dasar")

```
